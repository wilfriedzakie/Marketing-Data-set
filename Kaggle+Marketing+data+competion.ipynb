# Kaggle Marketing data competition

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn import metrics
import pandas as pd
import csv
from sklearn.metrics import cohen_kappa_score
```

## Load training and testing data

```python
train_file='C:/Users/zakie/Downloads/kaggle/train.csv'
df = pd.read_csv(train_file, sep=';')

test_file='C:/Users/zakie/Downloads/kaggle/test.csv'
test_df=pd.read_csv(test_file, sep=';', header=None)


#Target Data extraction
data_target=df['y']
data=df.drop('y', axis=1) 

print(data.shape)
print(data_target.shape)

from sklearn.model_selection import train_test_split
```

## Convert categorical data to numerical data 

```python
def cat_data(df):
    fields=df.columns.values
    
    for column in fields:
        text_values={}
        def convert_to_int(val):
            return text_values[val]
        
        if df[column].dtype!=np.int64 and df[column].dtype!=np.float64:
            value=df[column].values.tolist()
            unique_values=set(value)
            x=0
            for unique in unique_values:
                if unique not in text_values:
                    text_values[unique]=x
                    x+=1
            df[column]=list(map(convert_to_int,df[column]))
    return df
data=cat_data(data)
```

## Spliting data set into training and testing data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( data, data_target, test_size=0.4, random_state=0)
```

## AdaBoost Classifier training

```python
from sklearn.ensemble import AdaBoostClassifier
clf1 = AdaBoostClassifier(n_estimators=100)

clf1.fit(X_train,y_train)
y_pred3=clf1.predict(X_test)
print(cohen_kappa_score(y_test , y_pred3))
```

## Load and predict of Test Data 

```python
test_data= cat_data(test_df)
test_data.head()
test=clf1.predict(test_data)
```

## Convert data frame into np.array 

```python
predic1=[]
for i in test:
    if i=="no":
        predic1.append(0)
    else:
        predic1.append(1)

print(predic1)
```
